```python
import pandas as pd
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import cv2
import os
import json
import yaml
import shutil
import pydicom
from tqdm import tqdm
```

```python
df = pd.read_csv('/workspace/data/df_train_study_level_npy640_2_w_bbox.csv')
df.head()
```

```python
df_image = df.groupby('npy_path').first().reset_index()
```

```python
for i in tqdm(range(len(df_image))):
    row = df_image.iloc[i]
    new_image_path = row.npy_path.replace('train_640_2', 'yolo_data/images')
    shutil.copy(row.npy_path, new_image_path)
    name = new_image_path.split('/')[-1].replace('.png', '')
    label_path = f"/workspace/data/yolo_data/labels/{name}.txt"

    img = cv2.imread(row.npy_path)
    h, w, _ = img.shape

    bbox_df = df[df.npy_path == row.npy_path]
    bbox_text = ""
    if row.have_box:
        for n_box in range(len(bbox_df)):
            box = bbox_df.iloc[n_box]
            # yolo style bbox
            center_x = (box["x"] + box["w"]//2)/w
            center_y = (box["y"] + box["h"]//2)/h
            width = box["w"]/w
            height = box["h"]/h

            bbox_text += f'0 {center_x} {center_y} {width} {height}\n'


    with open(label_path, "w") as f:
        f.write(bbox_text)
```
