```python
import pandas as pd
import numpy as np
import ast
```

```python
df_bbox = pd.read_csv('/workspace/data/df_train_study_level_npy640_2_w_bbox.csv')
```

```python
df_bbox
```

```python
df_image_unique = df_bbox.groupby('id').first().reset_index()
```

```python
study_value_counts = df_image_unique.groupby('study_id')["image_id"].count().sort_values()
```

```python
many_image_index = study_value_counts[study_value_counts > 1].index
```

```python
df_bbox[df_bbox.study_id.isin(many_image_index)].sort_values('study_id').head(50)
```

```python
df_have_box_min_max = df_bbox.groupby('study_id').agg({"have_box": ["min", "max"]}).reset_index()["have_box"]
df_have_box_min_max.index = df_bbox.groupby('study_id').first().index
df_have_box_min_max["drop_zero_index"] = (df_have_box_min_max.sum(1) == 1) * 1
```

```python
drop_zero_index = df_have_box_min_max[df_have_box_min_max["drop_zero_index"] == 1].index
```

```python
df_drop = df_bbox[df_bbox.study_id.isin(drop_zero_index)]
df_others = df_bbox[~df_bbox.study_id.isin(drop_zero_index)]
```

```python
df_droped = df_drop[df_drop.have_box == 1]
```
