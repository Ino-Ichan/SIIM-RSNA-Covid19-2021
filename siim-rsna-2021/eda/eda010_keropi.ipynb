```python
def read_list_from_file(list_file, comment='#'):
    with open(list_file) as f:
        lines  = f.readlines()
    strings=[]
    for line in lines:
        if comment is not None:
            s = line.split(comment, 1)[0].strip()
        else:
            s = line.strip()
        if s != '':
            strings.append(s)
    return strings
```

```python
df_fold = pd.read_csv('/workspace/data/keropi/df_fold_rand830.csv')
df_duplicated = read_list_from_file('/workspace/data/keropi/duplicate.txt')
```

```python
df_fold.head(2)
```

```python
df_bbox2 = df_bbox.merge(df_fold, on="study_id", how="left")
```

```python
df_bbox3 = df_bbox2[~df_bbox2["image_id"].isin(df_duplicated)].reset_index(drop=True)
```

```python
print(df_bbox2.shape)
print(df_bbox3.shape)
```

```python
df_bbox3["is_none"] = 1 - df_bbox3["have_box"]
```
