```python
import pandas as pd
import numpy as np
import pydicom
import cv2
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/workspace/data/df_train_study_level.csv')
```

```python
path_list = df.dicom_path.values
```

```python
dicom = pydicom.read_file(path_list[0])
```

```python
image = dicom.pixel_array
```

```python
image.shape
```

```python
image.max()
```

```python
image.min()
```

```python
plt.figure()
plt.hist(image.flatten(), bins=200)
plt.show()
```

```python
plt.figure()
for i in range(30):
    image = pydicom.read_file(path_list[i]).pixel_array
    plt.hist(image.flatten(), bins=200, alpha=0.1)
plt.show()
```

```python
plt.figure()
for i in range(30):
    image = pydicom.read_file(path_list[i]).pixel_array
    image = (image - image.min()) / (image.max() - image.min())
    plt.hist(image.flatten(), bins=200, alpha=0.3)
plt.show()
```

```python
plt.figure()
for i in range(30):
    image = pydicom.read_file(path_list[i]).pixel_array
    image = (image - image.min()) / (image.max() - image.min())
    image = (image - image.mean()) / image.std()
    plt.hist(image.flatten(), bins=200, alpha=0.3)
    
# plt.ylim([0, 1e-5])
plt.show()
```

```python
plt.figure()
for i in range(30):
    image = pydicom.read_file(path_list[i]).pixel_array
    image = (image - image.min()) / (image.max() - image.min())
    image = (image - image.mean()) / image.std()
    plt.hist(image.flatten(), bins=200, alpha=0.3)
    
plt.ylim([0, 1e6])
plt.show()
```
