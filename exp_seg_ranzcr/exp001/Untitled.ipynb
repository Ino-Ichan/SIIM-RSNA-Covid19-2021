```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
df = pd.read_csv('/workspace/data/external/ranzcr/RANZCR_CLiP_lung_contours.csv')
```

```python
"/workspace/data/external/ranzcr/train/"
```

```python
img_root = "/workspace/data/external/ranzcr/train/"
lung_root = "/workspace/data/external/ranzcr/train_lung_masks/train_lung_masks/"

df["img_path"] = [os.path.join(img_root, i) + ".jpg" for i in df.StudyInstanceUID]
df["mask_path"] = [os.path.join(lung_root, i) + ".jpg" for i in df.StudyInstanceUID]
```

```python
for i in range(len(df)):
    row = df.iloc[i]
    if os.path.exists(row.img_path) == False:
        print(row.img_path)
    if os.path.exists(row.mask_path) == False:
        print(row.mask_path)
```

```python
plt.figure(figsize=(16, 8))
img = cv2.imread(row.img_path)
mask = cv2.imread(row.mask_path)
# red mask
mask[:,:,1:] = 0

plt.subplot(1, 3, 1)
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.imshow(mask)

plt.subplot(1, 3, 3)
plt.imshow(img)
plt.imshow(mask, alpha=0.3)

plt.show()
```
